```python
from transformers import GPT2LMHeadModel, GPT2TokenizerFast

MODEL_PATH = '../model_hub/gpt2-xl'
model = GPT2LMHeadModel.from_pretrained(MODEL_PATH).to('cuda')
tokenizer = GPT2TokenizerFast.from_pretrained(MODEL_PATH)
```

```python
import torch


be = tokenizer(['我喜欢中国'], return_tensors='pt')
input_ids, attention_mask = be['input_ids'].cuda(), be['attention_mask'].cuda()
output_ids = model.generate(input_ids=input_ids, attention_mask=attention_mask, do_sample=True, top_p=0.7, temperature=1, max_new_tokens=100, num_return_sequences=4)
tokenizer.batch_decode(output_ids, skip_special_tokens=True)
```
